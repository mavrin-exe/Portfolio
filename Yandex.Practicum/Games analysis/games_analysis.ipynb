#### Оценки пользователей

При построении гистограммы на первом этапе можно обратить внимание на отсутствие таковой по столбцу  `'user_score'`. Что странно, т.к. в таблице столбцы `'user_score'` и `'critic_score'` носят один и тот же характер - оценка игры в виде дробного числа. Тип данных object, вероятнее всего, был присвоен из-за наличия среди всех значений столбца некого выбивающегося, строкового. Проверим это:

```python
df['user_score'].unique()
```

```python
df.loc[df['user_score'] == 'tbd']['user_score'].count()
```

Среди оценок есть 2424 строк со значением `'tbd'` - от английского To Be Determined. Так обычно в каталогах обозначают  фильмы или игры, у которых ещё недостаточно оценок от пользователей для вычисления средней оценки. Причиной этому может быть как недавний релиз продукта, так и низкая популярность / нишевость. 

Строковое значение мешает нам использовать столбец в анализе, поэтому рассмотрим варианты решений:
1. **Удаление строк**. Не подойдёт, потеряем слишком много данных.
2. **Заполнение медианой / средним**. У нас нет на то каких-либо логичных оснований. Мы не сможем привязать медиану оценки к прочим столбцам, т.к. параметров слишком много и они чаще не связаны. Это будут синтетические данные, не имеющие отношения к реальности.
3. **Заменить `'tbd'` на `NaN`**. Наиболее безопасный способ в нашей ситуации. Сохраним данные и сможем работать с цифрами.

Заменим значение `'tbd'` на `NaN` и переведём значения столбца в тип `float`:

```python
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
```

#### Оценки критиков и пользователей

Выведем кол-во пропусков в столбцах:

```python
print('Пропусков в столбце "critic_score":', len(df.loc[df['critic_score'].isna()]))
print('Пропусков в столбце "user_score":', len(df.loc[df['user_score'].isna()]))
```

Количество пропусков достаточно большое и разнится. Учитывая то, что база охватывает большой период и не объединяет мультиплатформенные игры в одну запись, причинами возникновения пропусков могут быть:
- Игра старая, по ней нет отзывов на актуальных агрегаторах оценок. 
- Игра мультиплатформенная, не для всех её версий выставлялись оценки. 
- Игра не очень популярная, и по ней нет рецензий критиков, либо они вовсе отсутствуют на агрегаторах. 
- Иная причина (ошибка выгрузки и пр.) 

С первой подгруппой мы ничего не сможем сделать. Этих оценок не существует, никакое расчётное значение (медиана, среднее) не будет являться корректным - в контексте влияния на продажи оценка сверхпопулярного 'Тетриса' в средние 68 баллов будет выглядеть странно.

Игры из второй подгруппы выходили на нескольких платформах, а оценки стоят не у всех версий. История игростроя знает немало случаев, когда оценки у версий для разных платформ имели кардинальные отличия, поэтому заполнять пропуски расчётными значениями мы не в праве. 

Третья и четвёртая группа представляют сложность по причинам, схожим с первой. Присвоение расчётных значений (например, медиана или среднее по жанру) без привязки к игре и конкретным данным приведёт к резким выбросам в распределении значений. Поэтому оставим пропуски как есть.

#### Год выпуска

В датафрейме 268 пропусков по столбцу `'year_of_release'`.

```python
df.query('year_of_release.isna()').sample(5)
```

Не зная год наверняка, нам будет тяжело заполнить пропуски, т.к. медианы и средние могут принять некорректные значения. Однако в списке есть игры с указанием года в названии самой игры. Попробуем использовать это и сохранить часть данных. 

Выведем строки, где в названии встречается "ноль" и пропуск в `'year_of_release'`:

```python
df.query('(name.str.find("0") != -1) and (year_of_release.isna())')
```

В большинстве случаев это спортивные игры. Для них почти всегда верно следующее:
- Год в названии пишут, если они выпускаются ежегодно
- Выпускаются они прошлым календарным годом. Т.е. FIFA 2010 выпускается осенью 2009, и так далее. 

Зная это, мы можем заполнить часть пропусков. 
Остальные пропуски удалим. 

```python
for row in df.loc[df['year_of_release'].isna()].index:
    
    if '02' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2001
    elif '03' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2002
    elif '04' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2003
    elif '05' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2004
    elif '06' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2005
    elif '07' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2006
    elif '08' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2007
    elif '09' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2008
    elif '11' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2010
    # отдельно для Tomb Rider
    elif '13' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 2013
    elif '99' in df.loc[row,'name']:
        df.loc[row,'year_of_release']  = 1999

    else:
        df = df.drop(labels = row,axis = 0)
```

Пропуски заполнены. Осталось привести тип данных в соответствие. Год может быть только целым числом:

```python
df['year_of_release'] = df['year_of_release'].astype('int64')
```

## Проверка гипотез

Нам предстоит проверить две гипотезы:
1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. Средние пользовательские рейтинги жанров Action и Sports разные.

### Гипотеза 1: Xbox One и PC 

Точная формулировка задачи: "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые"

__________________________

Метод `st.ttest_ind()` проверяет вероятность равенства средних двух независимых совокупностей, поэтому мы можем сформулировать нулевую и альтернативную гипотезу так:
- H_0: Средние пользовательские рейтинги платформы Xbox One = Средние пользовательские рейтинги платформы PC
- H_a: Средние пользовательские рейтинги платформы Xbox One ≠ Средние пользовательские рейтинги платформы PC

Зададим уровень статистической значимости *alpha = 0.05*

```python
# Выделим необходимые нам выборки
xbox_data = act_df.loc[(~act_df['user_score'].isna()) & (act_df['platform'] =='XOne')]['user_score']
pc_data = act_df.loc[(~act_df['user_score'].isna()) & (act_df['platform'] =='PC')]['user_score']
```

```python
alpha = 0.05 
results = st.ttest_ind(
    xbox_data,
    pc_data,
    equal_var = True
)

print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
    
```

Нулевая гипотеза не отвергнута. Это значит, что пользователи в среднем действительно ставят одинаковые оценки играм для Xbox One и PC.

### Гипотеза 2: жанры Action и Sports

Точная формулировка задачи: "Средние пользовательские рейтинги жанров Action и Sports разные"

__________________________

Метод `st.ttest_ind()` проверяет вероятность равенства средних двух независимых совокупностей, поэтому при формулировании нулевой и альтернативной гипотезы нам следует поменять знаки равенства:
- H_0: Средние пользовательские рейтинги жанра Action = Средние пользовательские рейтинги жанра Sport
- H_a: Средние пользовательские рейтинги жанра Action Xbox One ≠ Средние пользовательские рейтинги жанра Sport

Зададим уровень статистической значимости *alpha = 0.05*

```python
action_data = act_df.loc[(~act_df['user_score'].isna()) & (act_df['genre'] =='Action')]['user_score']
sports_data = act_df.loc[(~act_df['user_score'].isna()) & (act_df['genre'] =='Sports')]['user_score']
```

```python
alpha = 0.05 
results = st.ttest_ind(
    action_data,
    sports_data,
    equal_var = False
)

print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
    
```

Нулевая гипотеза отвергнута, сработала альтернативная гипотеза. Мы меняли знаки равенства, значит наше первоначальное предположение оказалось верно: пользователи в среднем действительно ставят разные оценки играм жанров Action и Sports.

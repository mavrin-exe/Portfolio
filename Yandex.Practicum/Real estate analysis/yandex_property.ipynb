#### Кухня и жилая площадь

Заменим пропуски в столбцах `'kitchen_area'` и `'living_area'`. Обычно эти данные вносят пользователи, но, видимо, поля не являются обязательными для заполнения. Автоматического заполнения алгоритмами сайта не предусмотрено, т.к. эти площади для разных квартир могут сильно отличаться. Однако мы можем посчитать, какую долю кухня и жилая площадь занимают в общей площади квартиры `'total_area'` в среднем по базе. С помощью этих долей и заполним недостающие значения.

В процессе обезопасим себя от возникновения аномальных значений, где площадь жилая и кухни окажутся больше разумных пределов в 90% от всей квартиры. 

```python
x = []
for ind in df.loc[df['kitchen_area'].isna()].index:
    df.loc[ind, 'kitchen_area'] = \
    df.loc[ind, 'total_area'] * (df['kitchen_area'].median() / df['total_area'].median())
    if (df.loc[ind, 'kitchen_area'] + df.loc[ind, 'living_area']) / df.loc[ind, 'total_area'] > 0.9:
        x.append(ind)
        
for ind in df.loc[df['living_area'].isna()].index:
    df.loc[ind, 'living_area'] = \
    df.loc[ind, 'total_area'] * (df['living_area'].median() / df['total_area'].median())
    if ((df.loc[ind, 'kitchen_area'] + df.loc[ind, 'living_area']) / df.loc[ind, 'total_area'] > 0.9) & (ind not in x):
        x.append(ind)
        
print(f'Получилось {len(x)} строк со значением более 0.9')
```

Можно попытаться придумать более точный алгоритм расчёта площадей, но процесс может стать излишне трудоёмким. Отсутствие 186 строк вряд ли сильно скажется на дальнейшем исследовании, поэтому удалим их:

```python
df = df.drop(labels = x,axis = 0)
```

```python
df['kitchen_area'].isna().sum()
```

```python
df['living_area'].isna().sum()
```

Пропуски устранены. 

_______________________

Посмотрим на распределение значений и, при необходимости, избавимся от аномалий. 

```python
df.boxplot(column=['kitchen_area', 'living_area'], figsize=(10,5))
df[['kitchen_area', 'living_area']].describe()
```

Диаграмма размаха помещает все кухни площадью более 17 м в выброс. Попробуем углубиться. 

```python
kit_liv = df.loc[:,['rooms','total_area','living_area','kitchen_area']]
kit_liv['other_space'] = kit_liv['total_area'] - kit_liv['kitchen_area'] - kit_liv['living_area']
kit_liv['delta'] = kit_liv['other_space'] / kit_liv['total_area']
kit_liv.describe()
```

Таблица показывает, что существуют квартиры с отрицательной разницей между `'total_area'` и суммой `'kitchen_area'` + `'living_area'`. Это означает, что при публикации продавцы вносили некорректные данные. Например, могли указать совмещённую гостинную с кухней как 'кухня', а потом отдельно добавить площадь комнаты в 'жилую'. И ещё множество подобных вариаций. 

Если принять, что прочая площадь (балконы, санузлы, коридоры, кладовые) должна составлять не менее 10% и не более 40%, то можем посчитать кол-во аномальных квартир. Добавим также подсчёт квартир без комнат и с жилой площадью менее 10 м. 

```python
print('Квартир с долей прочих помещений менее 10%:', kit_liv.query('delta < 0.1')['delta'].count()) 
print('Квартир с долей прочих помещений более 40%:', kit_liv.query('delta > 0.4')['delta'].count())
print('Квартир без комнат:', kit_liv.query('rooms == 0')['rooms'].count())
print('Квартир с жилой площадью менее 10 м:', kit_liv.query('living_area < 10')['living_area'].count())
```

Во всех этих объявлениях продавцы спутали данные, а восстановить их с точки зрения логики будет весьма сложно.  В данном случае проще исключить их из дальнейшего анализа. 

```python
try:
    df = df.drop(labels = kit_liv.query('delta < 0.1').index, axis = 0)
    df = df.drop(labels = kit_liv.query('delta > 0.4').index, axis = 0)
except:
    print(f'Все строки уже удалены')
```

```python
df = df.query('living_area >= 10')
df = df.query('rooms != 0')
```

```python
df.boxplot(column=['kitchen_area', 'living_area'], figsize=(10,5))
df[['kitchen_area', 'living_area']].describe()
```

В базе остались редкие выдающиеся значения по столбцам `'kitchen_area'` и `'living_area'`, но такие квартиры могут существовать на практике. #### Апартаменты

#### Этажи

В столбце `'floors_total'` 86 пропусков. Сайт позволяет не указывать общее кол-во этажей при публикации. Продавцы могут пропустить эту графу по невнимательности. Медианы и средние значения тут не помогут, и мы могли бы удалить эти строки, т.к. их мало. Но попробуем оставить и заполнить каким-либо значением исходя из этажа, на котором находится квартира. 

```python
df.loc[df['floors_total'].isna(), 'floor'].value_counts()
```

Среди пропусков встречаются квартиры на этажах от 1-го до 24-го. Чтобы не терять строки, разделим их на три группы - '9-этажный дом' для квартир на этажах до 9-го, '16-этажный дом' для квартир от 10-го до 16-го этажа, и '24-этажный дом' для остальных:

```python
df.loc[(df['floors_total'].isna()) & (df['floor'] <= 9), 'floors_total'] = 9
df.loc[(df['floors_total'].isna()) & (df['floor'] <= 16) & (9 < df['floor']), 'floors_total'] = 16
df.loc[(df['floors_total'].isna()) & (df['floor'] <= 24) & (16 < df['floor']), 'floors_total'] = 24
```

```python
df['floors_total'].isna().sum()
```

```python
df.boxplot(column='floors_total', figsize=(10,5))
```

```python
df.query('floors_total > 40')
```

```python
df.loc[2253, 'floors_total'] = 6
```

```python
df['floors_total'] = df['floors_total'].astype('int')
```

Напишем функцию, которая заменит различные вариации населённых пунктов на единое для каждого типа (поселок, садовое товарищество и т.п.):

```python
def city_func(row):
    
    if 'поселок городского типа' in row:
        return row.replace('поселок городского типа', 'поселок')
    elif 'посёлок городского типа' in row:
        return row.replace('посёлок городского типа', 'поселок')
    elif 'городской посёлок' in row:
        return row.replace('городской посёлок', 'поселок')
    elif 'городской поселок' in row:
        return row.replace('городской поселок', 'поселок')
    elif 'коттеджный посёлок' in row:
        return row.replace('коттеджный посёлок', 'поселок')
    elif 'коттеджный поселок' in row:
        return row.replace('коттеджный поселок', 'поселок')
    elif 'садоводческое некоммерческое товарищество' in row:
        return row.replace('садоводческое некоммерческое товарищество', 'садовое товарищество')
    elif 'садоводческое товарищество' in row:
        return row.replace('садоводческое товарищество', 'садовое товарищество')
    elif 'посёлок при железнодорожной станции' in row:
        return row.replace('посёлок при железнодорожной станции', 'поселок')
    elif 'поселок при железнодорожной станции' in row:
        return row.replace('поселок при железнодорожной станции', 'поселок')
    elif 'поселок станции' in row:
        return row.replace('поселок станции', 'поселок')
    elif 'посёлок станции' in row:
        return row.replace('посёлок станции', 'поселок')
    #для "ё" в названиях:
    elif 'ё' in row:
        return row.replace('ё', 'е')
    
    else:
        return row
```

```python
df['locality_name'] = df['locality_name'].apply(city_func)
len(df['locality_name'].unique())
```

- `'per_meter'` - цена за 1 кв.м.;

- `'km_to_center'` - расстояние до центра города в километрах;

2. Напишем функцию, которая присваивает квартире категорию по признаку этажа, и применим её:

```python
# Функция:
def floor_type(row):
    try:
        if row['floor'] == 1:
            return 'первый'
        elif row['floor'] == row['floors_total']:
            return 'последний'
        elif (row['floor'] > 1) and (row['floor'] < row['floors_total']):
            return 'прочие'
    except:
        print('В столбце обнаружены ошибочные значения')

# Применим к главному датафрейму:    
df['floor_type'] = df.apply(floor_type, axis=1)
```

**Скачок на 3-м и 5-м километрах**

```python
spb_km.query('km_to_center == 3')['floor_type'].hist(
    bins=10, 
    figsize=(15, 5), 
    alpha=1, 
    legend=True);
spb_km.query('km_to_center == 5')['floor_type'].hist(
    bins=10, 
    figsize=(15, 5), 
    alpha=0.5, 
    legend=True).set_title('Кол-во квартир по типам этажей (на 3-м и 5-м км)', fontsize = 12, color = 'Black');
```

По гистограммам видна большая разница в кол-ве объявлений на "прочих" этажах. Как мы выяснили ранее, там средняя стоимость квартир чуть выше. Разница в кол-ве объявлений на первом и последних этажах менее разительная. Посчитаем цифры подробнее.

Сгруппируем объявления из СПб по типу этажей, выделим нужные нам километры и посчитаем:
1. Кол-во объявлений в каждой "этажной" группе
2. Общее число объявлений на этом километре
3. Долю каждой группы в общем кол-ве квартир на данном километре

```python
fifth = (
    spb_km.query('km_to_center == 5')
    .groupby(['km_to_center', 'floor_type'])['floor_type']
    .agg(['count', lambda x: len(spb_km.query('km_to_center == 5')['floor_type'])])
    .reset_index()
)

fifth['of_all'] = round(fifth['count'] / fifth['<lambda_0>'], 2)
fifth
```

```python
third = (
    spb_km.query('km_to_center == 3')
    .groupby(['km_to_center', 'floor_type'])['floor_type']
    .agg(['count', lambda x: len(spb_km.query('km_to_center == 3')['floor_type'])])
    .reset_index()
)

third['of_all'] = round(third['count'] / third['<lambda_0>'], 2)
third
```

Для наглядности построим круговую диаграмму для обеих таблиц:

```python
third.plot(kind='pie', y='of_all', grid=True, figsize=(15, 4), label='3-й км', autopct='%1.1f%%', labels = third.floor_type);
plt.legend(fontsize=7, loc='best');

fifth.plot(kind='pie', y='of_all', grid=True, figsize=(15, 4), label='5-й км', autopct='%1.1f%%', labels = fifth.floor_type);
plt.legend(fontsize=7, loc='best');
```

**Наблюдения:**

- на 3-м километре первый и последний этажи суммарно занимают почти 40% от всех объявлений;
- на 5-м километре первый и последний этажи суммарно занимают чуть менее 25% от всех объявлений;
- эта разница объясняет пик и падение на графике зависимости цены от расстояния до центра.

**Вывод:**

Зная, что на первом и последнем этаже квартиры дешевле, чем на остальных, а также рассчитав доли, мы смогли объяснить пик и падение на графике зависимости цены от расстояния до центра для 3-го и 5-го километра.
